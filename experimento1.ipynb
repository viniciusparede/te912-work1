## 4) Gere o sinal x = 2 * t * 0.9^t com t = 0:49

```python
# Vetor tempo de 50 posições
t = np.linspace(0, 49, 50)

# Sinal
sinal = 2 * t * pow(0.9, t)
```

### a. Adicione ruído gaussiano no sinal (função randn)

```python
# Ruído Gaussiano
ruido = np.random.randn(1, 50).T
ruido_list = list()
for _, element in enumerate(ruido):
    ruido_list.append(element[0])
ruido = np.array(ruido_list)

# Sinal ruidoso
z = ruido + sinal
```

### b. Aplique o filtro de médio móvel de 3 pontos do sinal ruidoso

```python
def moving_average(a, n) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

filtro = moving_average(z, 3)
```

### c. Vizualize o sinal original, o sinal ruidoso e o sinal filtrado

```python
fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(15, 8))

ax[0].plot(t, sinal)
ax[1].plot(t, ruido)
ax[2].plot(t, z)
ax[3].plot(filtro)

fig.tight_layout()
```

## 6) A Transformada Rápidade Fourier (FFT) X[k] de uma sequência finita x[n] pode ser computada no Octave / Matlabusando a função fft. É possível usar a função na forma fft(x), gerando um sinal do mesmo tamanho de x, ou na forma fft(x,L) computando  a DFT  em L pontos.  Crie  uma sequência x  =  cos(2*pi*0.4*n) de tamanho  32.  Compute  e  visualizea DFT  desta sequência.  Repita  a  operação com L=8 para a DFT. Visualize usando stem a magnitude (abs) e a fase (angle) dos dois sinais obtidos.

```python
# Definindo o tamanho
n = np.linspace(0, 31, 32)

# Sinal
x = np.cos(2 * np.pi * 0.4 * n)

# DFT (Discret Fourier Transform) -> Tamanho 32
fft_32 = np.fft.fft(x, n=32)
modulo_32 = np.absolute(fft_32) 
fase_32 = np.angle(fft_32)

# DFT (Discret Fourier Transform) -> Tamanho 8
fft_8 = np.fft.fft(x, n=8)
modulo_8 = np.absolute(fft_8)
fase_8 =  np.angle(fft_8)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

axes[0][0].stem(n, modulo_32, basefmt='--')
axes[0][1].stem(n, np.rad2deg(fase_32), basefmt='--')
axes[0][0].set_ylabel("|Módulo DFT 32|", size=12)
axes[0][1].set_ylabel("Ângulo DFT 32 [°]", size=12)


axes[1][0].stem(modulo_8, basefmt='--')
axes[1][1].stem(np.rad2deg(fase_8), basefmt='--')
axes[1][0].set_ylabel("|Módulo DFT 8|", size=12)
axes[1][1].set_ylabel("Ângulo DFT 8 [°]", size=12)

plt.tight_layout()
```

## 7) Gere um sinal x que seja o somatório de duas senóides de 300 e 3000 Hz, com duração de 2 segundos e frequência de amostragem de 8kHz

```python
# Frequência de Amostragem = 8kHz
freq_amostragem = 8000

# Array de amostras (tempo)
tempo = np.arange(0, 2, 1/freq_amostragem)

# Sinal 1 -> Domínio do tempo
y1 = np.sin(2 * np.pi * 300 * tempo)

# Sinal 1 --> Domínio da frequência
y1_dft = np.fft.fft(y1, n=tempo.shape[0])
modulo_y1_dft = np.absolute(y1_dft)

# Sinal 2 - > Domínio do tempo
y2 = np.sin(2 * np.pi * 3e3 * tempo)

# Sinal 2 -> Domínio da frequência
y2_dft = np.fft.fft(y2, n=tempo.shape[0])
modulo_y2_dft = np.absolute(y2_dft)

# Sinal x -> Somátorio dos sinais 1 e 2 (Domínio do tempo)
x = y1 + y2

# Sinal x -> Domínio da frequência
x_fft = np.fft.fft(x, n=tempo.shape[0])
modulo_x_dft = np.absolute(x_fft)
```

### a. Mostre os gráficos do sinal no domínio do tempo, usando a função plot e a frequeência, usando a função stem

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))


axes[0][0].plot(tempo[0:100], y1[0:100])
axes[0][1].plot(tempo[0:100], y2[0:100])
axes[0][2].plot(tempo[0:100], x[0:100])

axes[1][0].stem(modulo_y1_dft[0:100], basefmt='--')
axes[1][1].stem(modulo_y2_dft[0:100], basefmt='--')
axes[1][2].stem(modulo_x_dft[0:100], basefmt='--')

axes[0][0].set_ylabel("Senóide 300 Hz (Tempo)", size=12)
axes[0][1].set_ylabel("Senóide 3 kHz (Tempo)", size=12)
axes[1][0].set_ylabel("Senóide 300 Hz (Tempo)", size=12)
axes[1][1].set_ylabel("Senóide 3 kHz (Frequência)", size=12)

fig.tight_layout()
```

## 8) O padrão DTMF (Dial Tone Multi Frequency) é amplamente utilizado em sistemas de telefonia. O padrão industrial de especificação de frequências para todas as teclas segue o diagrama da Figura 1:

```python
# Esse exercício está com um bug ou talvez algum conceito equivocado de PDS.
# O aúdio "audio_completed.wav" deve ser pelo menos parecido com "DTMF1.wav"
# Questionar isso para o professor na próxima aula
from scipy.io import wavfile
import IPython
import math

def matriz_de_frequencia (frequencias_altas: list, frequencias_baixas: list) -> list:
    frequency_list = list()
    for _, frequencia_baixa in enumerate(frequencias_baixas):
        for _, frequencia_alta in enumerate(frequencias_altas):
            append_values = [frequencia_baixa, frequencia_alta]
            frequency_list.append(append_values)
    return frequency_list

def to_signal (frequencies: list, tempo: np.ndarray) -> list:
    audio_list = [0] * 12
    amplitude = np.iinfo(np.int16).max

    for index, frequency in enumerate(frequencies):
        senoide_baixa = amplitude * np.sin(2 * np.pi * (frequency[0]) * tempo)
        senoide_alta = amplitude * np.sin(2 * np.pi * frequency[1] * tempo)
        audio = senoide_baixa + senoide_alta
        audio_list[index] = audio
    return audio_list

def recorder_audio (audio_list: list, sample_rate: float):
    for index, _ in enumerate(audio_list):
        if index == 9:
            filename = '*' + '_sound.wav'
            wavfile.write(filename, int(sample_rate), audio_list[index])
        elif index == 10:
            filename = str(0) + '_sound.wav'
            wavfile.write(filename, int(sample_rate), audio_list[index])
        elif index == 11:
            filename = '#' + '_sound.wav'
            wavfile.write(filename, int(sample_rate), audio_list[index])
        else:
            filename = str(index) + '_sound.wav'
            wavfile.write(filename, int(sample_rate), audio_list[index])

def play_audio(audio_list, sample_rate: float):
    audio_complete = list()
    for _, audio_sound in enumerate(audio_list):
        for index, _ in enumerate(audio_sound):
            audio_complete.append(audio_sound[index])
    wavfile.write('audio_completed.wav', rate=int(sample_rate), data=np.array(audio_complete).astype(np.float32))
    return audio_complete

freq_altas = [1209, 1336, 1477]
freq_baixas = [697, 770, 852, 941]

freq_amostragem = 8e3

frequencies =  matriz_de_frequencia(freq_altas, freq_baixas)
tempo = np.arange(0, 0.1, 1/freq_amostragem)
audio_list = to_signal(frequencies, tempo)

recorder_audio(audio_list, freq_amostragem)
audio = play_audio(audio_list, freq_amostragem)
IPython.display.Audio("audio_completed.wav")
```
